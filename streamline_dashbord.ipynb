```python
import streamlit as st
import pandas as pd
import seaborn as sns
import plotly.express as px

# Load the Iris dataset
@st.cache
def load_data():
    iris_data = sns.load_dataset('iris')
    return iris_data

# Function to display summary statistics
def display_summary(data):
    st.subheader("Summary Statistics")
    st.write(data.describe())

# Function to display count of each species
def display_species_count(data):
    st.subheader("Species Count")
    species_count = data['species'].value_counts()
    st.write(species_count)

# Function to display box plot
def display_box_plot(data):
    st.subheader("Box Plot")
    fig = px.box(data, y="sepal_length", x="species", points="all")
    st.plotly_chart(fig)

# Function to display scatter plot
def display_scatter_plot(data):
    st.subheader("Scatter Plot")
    fig = px.scatter(data, x="sepal_length", y="sepal_width", color="species")
    st.plotly_chart(fig)

# Function to display pair plot
def display_pair_plot(data):
    st.subheader("Pair Plot")
    fig = px.scatter_matrix(data, dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"], color="species")
    st.plotly_chart(fig)

# Main function to run the Streamlit app
def main():
    # Set app title and layout
    st.title("Iris Dataset Exploratory Data Analysis")
    st.sidebar.title("EDA Dashboard")
    st.sidebar.markdown("Select a plot to visualize")

    # Load the dataset
    iris_data = load_data()

    # Display summary statistics
    display_summary(iris_data)

    # Display count of each species
    display_species_count(iris_data)

    # Display box plot
    display_box_plot(iris_data)

    # Display scatter plot
    display_scatter_plot(iris_data)

    # Display pair plot
    display_pair_plot(iris_data)

# Run the app
if __name__ == '__main__':
    main()
```
